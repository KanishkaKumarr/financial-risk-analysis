```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned_transactions.csv")
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df.head()
```

```python
monthly_summary = (
    df
    .groupby(['YearMonth', 'TransactionType'])['TransactionAmount']
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)
```

```python
monthly_summary['Net_Transaction'] = (
    monthly_summary.get('deposit', 0)
    - monthly_summary.get('withdrawal', 0)
    - monthly_summary.get('payment', 0)
    - monthly_summary.get('transfer', 0)
)

monthly_summary.head()
```

```python
yearly_summary = (
    df
    .groupby(['Year', 'TransactionType'])['TransactionAmount']
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)
```

```python
yearly_summary['Net_Transaction'] = (
    yearly_summary.get('deposit', 0)
    - yearly_summary.get('withdrawal', 0)
    - yearly_summary.get('payment', 0)
    - yearly_summary.get('transfer', 0)
)

yearly_summary.head()
```
